```python
#设置超参数
seed =100
img_size=228
batch_size=75

train_dir = './archive/11_sets_Hispanics/'
valid_dir = './output/val/'
test_dir = './output/test/'
```

```python
#数据分析
class_names = []
class_count = []
train_examples = 0
for f in os.listdir(train_dir):
    files = gb.glob(pathname=str(train_dir + '/' + '/*'))
    class_names.append(f)
    class_count.append(len(files))
    train_examples += len(files)
plt.figure(figsize=(50,5))
sns.barplot(x=class_names,y=class_count)
plt.show()
```

```python
#读取数据
train_gen = ImageDataGenerator(
    rotation_range=25,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    preprocessing_function=tf.keras.applications.vgg19.preprocess_input,
    dtype=tf.float32
)
train_batch = train_gen.flow_from_directory(
    directory=train_dir,
    target_size=(img_size,img_size),
    batch_size=batch_size,
    class_mode='sparse',
    seed=seed
)
#验证数据
valid_batch = train_gen.flow_from_directory(
    directory=valid_dir,
    target_size=(img_size,img_size),
    batch_size=batch_size,
    class_mode='sparse',
    seed=seed
)
```

```python
#查看数据
imgs,labels = next(train_batch)
i = 0
plt.figure(figsize=(20,20))
for img,label in zip(imgs,labels):
    plt.subplot(3,3,i+1)
    plt.imshow(img.astype('uint8'))
    plt.xlabel(class_names[label.astype('int32')])
    plt.yticks([])
    plt.xticks([])
    i+=1
    if i ==9:
        break
plt.tight_layout()   #调整多图，单图标签，只考虑刻度标签，轴标签，标题
#关键字 ：pad=0.4,w_pad=0.4,h_pad=0.4
plt.show()
```

```python
#准备模型
image_shape = (img_size,img_size) + (3,)
base_mode = tf.keras.applications.VGG19(
    input_shape=image_shape,
    include_top=False,
    weights='imagenet')
base_mode.summary()
#base_model.trainable 若不消除，则会出现精度非常低，训练不上去的等问题
base_mode.trainable = False#设置预训练为否
```

```python
inputs = tf.keras.Input(shape=(image_shape))
x = base_mode(inputs,training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(512,activation='relu')(x)
x = layers.Dense(512,activation='relu')(x)
x = layers.Dropout(0.1)(x)
outputs = layers.Dense(11,activation='softmax')(x)
model = tf.keras.Model(inputs,outputs)
model.summary()
```

```python
cp_callback = tf.keras.callbacks.ModelCheckpoint('training_cp/cp.ckpt',save_weights_only=True)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy']
             )
history = model.fit(train_batch,
                    epochs=50,
                    steps_per_epoch=165//batch_size,
                   validation_data=valid_batch,
                   validation_batch_size=33//batch_size)
```

```python
model.save('cp_model.h5')
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'],'o-',label='train accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend(loc='lower right')

plt.subplot(122)
plt.plot(history.history['loss'],'o-',label='train loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```
